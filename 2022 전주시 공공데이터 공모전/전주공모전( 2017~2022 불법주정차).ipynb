```python
import pandas as pd
import numpy as np

df = pd.read_csv("D:/경기대/2학년_2학기/dna 동아리/전주공모전/전라북도 전주시_불법주정차 단속현황_20220830.csv", encoding='cp949')
df.head()
```

```python
print(df.info())
print(df.shape)
```

```python
df[df['단속장소명'].isnull()]
```

```python
df = df.dropna(axis=0)
df.info()
```

```python
df['월'] = df['단속일자'].str.split("-", expand=True)[1]
df['일'] = df['단속일자'].str.split("-", expand=True)[2]
df.head()
```

```python
df['시간'] = df['단속시간'].str.split(":", expand=True)[0]
df['분'] = df['단속시간'].str.split(":", expand=True)[1]

df.head()
```

```python
df = df.astype({'시간':'int32', '월' : 'int32', '일' : 'int32', '분' : 'int32'})
df.info()
```

```python
df1 = df[df['시간'] <= 11].copy()
df1['오전/오후'] = '오전'

df2 = df[df['시간'] > 11].copy()
df2['오전/오후'] = '오후'

result = pd.concat([df1,df2])
result
```

```python
#요일컬럼 생성을 위한 단속일자 type 바꾸기
result['단속일자_datetime'] = pd.to_datetime(result['단속일자'])
result.info()
```

```python
result['요일'] = result['단속일자_datetime'].dt.weekday

weekday_list = ['월', '화', '수', '목', '금', '토', '일']
result['요일'] = result.apply(lambda x : weekday_list[x['요일']], axis = 1)
```

```python
result['지자체명'] = '전라북도 전주시'
```

```python
#띄어쓰기를 위한 공백 추가 
result['공백'] = ' '
result['상세주소'] = result['지자체명'] + result['공백'] + result['구청구분'] + result['공백'] + result['단속장소명']
```

```python
# 상세시간 생성을 위한 type 통일
result = result.astype({'단속일자' : 'object'})
result.info()
```

```python
time = ['단속일자', '단속시간']
result['상세시간'] = result[time].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

a = ['단속장소명','상세주소', '상세시간', '단속일자', '단속시간', '요일', '오전/오후', '구청구분', '단속구분']
result = result[a]
result
```

```python
result['단속일자'].value_counts()
```

```python
result2 = result['상세주소'].value_counts().to_frame()
result2 = result2.reset_index().rename(columns = {'index':'주소', '상세주소':'불법주차수'})

result2
```

```python
result_top = result['상세주소'].value_counts().loc[lambda x: x >= 100].to_frame()
result_top = result_top.reset_index().rename(columns = {'index':'주소', '상세주소':'불법주차수'})
result_top
```

```python
result_top['주소']= result_top['주소'].str.replace(pat=r'[^\w]',repl=r' ',regex=True) 
#pat: 찾고자 하는 정규표현식
#특수문자만 나오게하는 정규표현식 r'[^\w]'
#repl: 대체하고자 하는 문자
result_top
```

```python
result2 = result['상세주소'].value_counts().to_frame()
result2 = result.reset_index().rename('index':'주소', '상세주소':'불법주차수')
result2
```

```python
result2.to_csv('전주불법주차위치.csv', index=False, encoding='cp949')
result.to_csv('전주불법주차.csv', index=False, encoding='cp949')
```
