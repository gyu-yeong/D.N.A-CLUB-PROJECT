### 단속일자 월, 일로 나누기

#### 1. " "안의 것 기준으로 스플릿
#### 2. expand 함수를 True로 둬 df형태로 만들기
#### 3. 만든 정보를 기존 df 안에 담기

```python
df['단속일자'].str.split("-", expand=True) #expand 함수는 df형태로 변환해줌
```

```python
df['월'] = df['단속일자'].str.split("-", expand=True)[1]
df['일'] = df['단속일자'].str.split("-", expand=True)[2]
df.head()
```

```python
df.tail()
```

### 단속시간 기준으로 오전 오후 나누기

```python
df['단속시간'].str.split(":", expand=True)
```

```python
df['시간'] = df['단속시간'].str.split(":", expand=True)[0]
df['분'] = df['단속시간'].str.split(":", expand=True)[1]

df.head()
```

```python
df.info()
```

```python
df = df.astype({'시간':'int32', '월' : 'int32', '일' : 'int32', '분' : 'int32'})
df.info()
```

```python
df1 = df[df['시간'] <= 11].copy()
df1['오전/오후'] = '오전'
```

```python
df2 = df[df['시간'] > 11].copy()
df2['오전/오후'] = '오후'
```

```python
result = pd.concat([df1,df2])
result
```

```python
result.info()
```

### 요일 컬럼 생성

```python
#요일컬럼 생성을 위한 단속일자 type 바꾸기
result['단속일자_datetime'] = pd.to_datetime(result['단속일자'])
result.info()
```

```python
result['요일'] = result['단속일자_datetime'].dt.weekday

weekday_list = ['월', '화', '수', '목', '금', '토', '일']
result['요일'] = result.apply(lambda x : weekday_list[x['요일']], axis = 1)
```

```python
result.head()
```

```python
#띄어쓰기를 위한 공백 추가 
result['공백'] = ' '

result['상세주소'] = result['지자체명'] + result['공백'] + result['구청구분'] + result['공백'] + result['단속장소명']
```

### 상세시간 컬럼 생성하기 (단속일자 + 단속시간 + 요일)

```python
# 상세시간 생성을 위한 type 통일
result = result.astype({'단속일자' : 'object'})
result.info()
```

```python
time = ['단속일자', '단속시간']
```

```python
result['상세시간'] = result[time].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
```

```python
result.head()
```

## 데이터프레임 정리 이후, csv로 저장하기 

```python
a = ['단속장소명','상세주소', '상세시간', '단속일자', '단속시간', '요일', '오전/오후', '구청구분', '단속구분']
result = result[a]
result
```

```python
#구만 뽑아보기
result['동 정보'] = result['상세주소'].str.split(' ').str[3]
result
```
