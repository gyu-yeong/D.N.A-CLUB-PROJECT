```python
#워드클라우드 및 명사 분류 함수
def df2str(df):
    s = [s for s in df]
    document = ""
    for i in range(len(s)):
        document += s[i]
    return document
    
def get_noun(text):
    okt = Okt()
    noun = okt.nouns(text)
    for i,v in enumerate(noun):
        if len(v) < 2:
            noun.pop(i)
    count = Counter(noun)
    noun_list = count.most_common(100)
    
    return noun_list

def visualize(noun_list, title):
    wc = WordCloud(
    font_path = "malgun.ttf",     
    background_color = 'white',
    colormap = 'Dark2',
    width = 800,
    height = 800).generate_from_frequencies(dict(noun_list))
    
    plt.figure(figsize=(10,10))
    plt.suptitle("Word Cloud",fontsize=40)
    plt.title(title, fontsize=20)
    plt.imshow(wc, interpolation = "lanczos")
    plt.axis("off")
    plt.show()
    
    return wc
```

```python
document = df2str(df_name)
noun_list = get_noun(document)
noun_list[0:6]
```

```python
total_name = visualize(noun_list,"df_name")
```
