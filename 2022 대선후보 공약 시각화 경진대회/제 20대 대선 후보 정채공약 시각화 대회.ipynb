👩‍💼결론에서 후보 추천 기준은 아래의 각 4가지 부분의 점수의 합계입니다.👩‍💼

1. **공공성** - 복지,교육 및 문화 
국민의 인간다운 삶을 위해 사회 복지 서비스는 사회 구성원 전체와 관련되는 성격을 가진 복지와 교육문화 카테고리를 공공성으로 지정하였습니다.

2. **성장성** - 경제,일자리,코로나
일자리와 경제는 국가 경제 성장에 필수불가결요소입니다.
코로나 지원금 역시도 경제적 지원이 강한 요소임으로 성장성으로 지정하였습니다.

3. **지속가능성** - 과학,기후위기 
기후위기와 과학은 인간과 자원의 공생, 4차산업 혁명 대비 차세대 과학기술에 관하므로 지속가능성으로 지정하였습니다.

4. **안보성** - 안보및외교
외교 및 안보는 본래의 의미대로 안보성으로 지정하였습니다.


🤵‍♂️점수부여 기준:
*   평균이상 값 : 3점
*   중앙값이상 : 2점 
*   중앙값미만 : 1점

따라서 공공성/성장성/안보성/지속가능성 카테고리를 생성한것을 토대로 사용자가 점수를 부여하여 관심 공약을 선택 할 수 있습니다.
