```python
# 연령 컬럼들의 범위
age_columns = ['계0~4세', '계5~9세', '계10~14세', '계15~19세', '계20~24세', '계25~29세', '계30~34세',
               '계35~39세', '계40~44세', '계45~49세', '계50~54세', '계55~59세', '계60~64세', '계65~69세',
               '계70~74세', '계75~79세', '계80~84세', '계85~89세', '계90~94세', '계95~100세']

# 연령 컬럼들을 기반으로 중위연령을 계산하는 함수
def calculate_median_age(row):
    age_values = []
    for age_col in age_columns:
        age_range = age_col.split('~')  # 연령 범위를 분리
        age_min = int(age_range[0].replace('계', ''))  # 연령 범위의 최소값
        age_max = int(age_range[1].replace('세', ''))  # 연령 범위의 최대값
        age_count = row[age_col]  # 해당 연령 범위의 인원수
        age_values.extend([age_min] * age_count)  # 인원수를 연령값으로 변환하여 리스트에 추가
    median_age = np.median(age_values)
    return median_age

# 행정동별로 중위연령을 계산
df1['중위연령'] = df1.apply(calculate_median_age, axis=1)
```

### 1세 단위 인구데이터로 중위연령 계산

```python
df2 = pd.read_csv(url+'연령별인구(1세단위).csv',encoding='cp949')
```

```python
df2.head(2)
```

```python
# '행정구역' 열의 괄호 문자 제거
df2['행정구역'] = df2['행정구역'].apply(remove_parentheses)

# 컬럼명 수정
new_columns = [df2.columns[0]] + [col.split('_')[1] + col.split('_')[2] for col in df2.columns[1:]]
df2.columns = new_columns

a = df2['행정구역'].str.split()
df2['행정구역'] = a.apply(lambda x: x if len(x) >= 4 else None)
df2 = df2.dropna(subset=['행정구역'])

df2['행정구역'] = df2['행정구역'].apply(lambda x: ' '.join(x))

df2['자치구'] = df2['행정구역'].str.split().apply(lambda x: x[2])

cols = list(df2.columns)
cols.insert(1,cols.pop(-1))

df2 = df2[cols]
```

```python
df2 = df2.replace(',', '', regex=True) # ,쉼표 표시 제거해야댐 계산하려면
df2.iloc[:, 2:] = df2.iloc[:, 2:].apply(pd.to_numeric) #수치형 변환도해야댐

df2['읍면동'] = df2['행정구역'].str.split().apply(lambda x: x[3])
df2 = df2.reset_index(drop=True)
df2 = df2.rename(columns={'읍면동': '행정동'})

df2 = df2.rename(columns={'계100세 이상': '계100세'})
```

```python
df2.head(1)
```

```python
# 중위연령을 계산하기 위한 함수 정의
def calculate_median_age(row):
    age_values = []
    for i in range(101):
        age = '계{}세'.format(i)
        age_values.extend([i] * row[age])
    return pd.Series({'중위연령': pd.Series(age_values).median()})
```

```python
# 중위연령을 구하기 위해 행정동과 계0세부터 계100세까지의 컬럼 선택
df_age = df2[['행정동'] + ['계{}세'.format(i) for i in range(101)]]

# 각 행정동별로 중위연령 계산
median_age = df_age.apply(calculate_median_age, axis=1)
```

```python
df2['중위연령'] = median_age
```

```python
df2.to_csv('진짜중위연령.csv',encoding='cp949')
```
