## 대여스테이션, 반납스테이션 0 값인 행 제거

- 이유: 0은 대여소가 아닌 곳이라고 판단

```python
tasu_2021.drop(tasu_2021.loc[tasu_2021['대여스테이션']== 0].index, inplace = True)
```

```python
tasu_2021.drop(tasu_2021.loc[tasu_2021['반납스테이션']== 0].index, inplace = True)
```

```python
station1=station.rename(columns = {'번호':'대여스테이션'})
```

```python
total1 = pd.merge(tasu_2021, station1, how = 'outer', on='대여스테이션')
```

```python
station2=station1.rename(columns = {'대여스테이션':'반납스테이션'})
```

```python
total = pd.merge(total1, station2, how = 'outer', on= '반납스테이션')
```

```python
total
```

```python
total=total.drop(['대여 위도'], axis = 1)
total=total.drop(['대여 경도'], axis = 1)
total=total.drop(['반납 위도'], axis = 1)
total=total.drop(['반납 경도'], axis = 1)
```

* 대여 반납 위도 경도 삭제

```python
total=total.rename(columns = {'정류장_x':'대여장소'})
total=total.rename(columns = {'위치_x':'대여위치'})
total=total.rename(columns = {'정류장_y':'반납장소'})
total=total.rename(columns = {'위치_y':'반납위치'})
total=total.rename(columns = {'Latitude_x':'대여위도'})
total=total.rename(columns = {'Longitude_x':'대여경도'})
total=total.rename(columns = {'Latitude_y':'반납위도'})
total=total.rename(columns = {'Longitude_y':'반납경도'})
```
