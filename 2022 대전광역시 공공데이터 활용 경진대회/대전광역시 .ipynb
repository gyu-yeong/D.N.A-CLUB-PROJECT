```python
import pandas as pd
df=pd.read_csv("C:/Users/user/Desktop/통합 문서1.CSV", encoding='euc-kr')
df
```

```python
dongu1 = df['타슈 시군구'] == '동구'
dongu2 = df['대중교통 시군구'] == '동구'
dongu = df[dongu1 & dongu2]
dongu
```

```python
daedukgu1 = df['타슈 시군구'] == '대덕구'
daedukgu2 = df['대중교통 시군구'] == '대덕구'
daedukgu = df[daedukgu1 & daedukgu2]
```

```python
yougu1 = df['타슈 시군구'] == '유성구'
yougu2 = df['대중교통 시군구'] == '유성구'
youkgu = df[yougu1 & yougu2]
```

```python
seogu1 = df['대중교통 시군구'] == '서구'
seogu2 = df['타슈 시군구'] == '서구'
seogu = df[seogu1 & seogu2]
```

```python
jungu1 = df['대중교통 시군구'] == '중구'
jungu2 = df['타슈 시군구'] == '중구'
jungu = df[jungu1 & jungu2]
```

```python
result1 = pd.concat([dongu, daedukgu, youkgu, seogu, jungu])
df1 = pd.DataFrame(result1)
```
